## 3. Prepare Phase
### 3.1 Data Overview 
This dataset contains minute-level fitness tracker data collected from thirty three eligible Fitbit users who participated in a survey distributed via Amazon Mechanical Turk between December 3, 2016, and December 5, 2016. Each respondent consented to share their personal tracker data, which includes detailed information on physical activity, heart rate, and sleep monitoring

The dataset is organized so individual reports can be accessed using the export session ID (column A) or timestamp (column B). Variations in the output data are attributed to different types of Fitbit trackers, as well as individual user behaviors and preferences in tracking their fitness activities.

This dataset, titled "FitBit Fitness Tracker Data," is made available under the CC0: Public Domain license through Mobius on Kaggle. This open-access status makes it a valuable resource for analyzing fitness patterns and health metrics.


The datasets represent physical activity, heart rate, and sleep data for 33 Fitbit users. They span two periods: March 12, 2016, to April 12, 2016, and April 12, 2016, to May 12, 2016. The data is spread across various CSV files. Below is an optimized data organization and an explanation of the columns in these datasets.

---

#### 1. `dailyActivity_merged`

This dataset contains daily activity summaries for the users.

| **Column Name**               | **Description**                                           |
| ----------------------------- | --------------------------------------------------------- |
| `Id`                          | User identifier (unique per person)                       |
| `ActivityDate`                | Date of the activity in `MM/DD/YYYY` format               |
| `TotalSteps`                  | Total number of steps taken on that day                   |
| `TotalDistance`               | Total distance covered in miles                           |
| `TrackerDistance`             | Distance tracked by the Fitbit device                     |
| `LoggedActivitiesDistance`    | Distance covered during logged activities                 |
| `VeryActiveDistance`          | Distance covered while very active                        |
| `ModeratelyActiveDistance`    | Distance covered while moderately active                  |
| `LightActiveDistance`         | Distance covered while lightly active                     |
| `SedentaryActiveDistance`     | Distance covered while sedentary                          |
| `VeryActiveMinutes`           | Minutes spent being very active                           |
| `FairlyActiveMinutes`         | Minutes spent being fairly active                         |
| `LightlyActiveMinutes`        | Minutes spent being lightly active                        |
| `SedentaryMinutes`            | Minutes spent in a sedentary state                        |
| `Calories`                    | Total calories burned during the day                      |

---

#### 2. `hourlySteps_merged`

This dataset contains hourly step counts for each user.

| **Column Name**    | **Description**                                           |
| ------------------ | --------------------------------------------------------- |
| `Id`               | User identifier (unique per person)                       |
| `ActivityHour`     | Timestamp of the activity in `MM/DD/YYYY HH:MM:SS AM/PM` format |
| `StepTotal`        | Total number of steps taken in that hour                  |

---

#### 3. `hourlyIntensities_merged`

This dataset contains hourly intensity levels of activities.

| **Column Name**      | **Description**                                           |
| -------------------- | --------------------------------------------------------- |
| `Id`                 | User identifier (unique per person)                       |
| `ActivityHour`       | Timestamp of the activity in `YYYY-MM-DD HH:MM:SS` format |
| `TotalIntensity`     | Total intensity of activities in that hour                |
| `AverageIntensity`   | Average intensity of activities in that hour              |
| `weekDay`            | The day of the week (e.g., Monday, Tuesday)               |
| `time`               | The hour of the day in `HH:MM:SS` format                  |

---

### 3.1.1 Data Characteristics:
- **Daily Activity**: Summarized data for each day, including steps, distances, and calories burned.
- **Hourly Steps**: Detailed step count data for each hour of the day.
- **Hourly Intensities**: Intensity metrics recorded hourly, representing the user's physical activity level.



### 3.1.2 Exclusion of Other Datasets

Some datasets were excluded from the analysis because they contained data from only 8 users, which is not a representative sample size for drawing meaningful insights. Since the datasets with 33 users provide a more comprehensive view of the activity trends, only those datasets (`dailyActivity_merged`, `hourlySteps_merged`, and `hourlyIntensities_merged`) were used for this analysis. Including datasets with fewer participants could introduce bias or lead to less reliable conclusions.


### 3.2  Data Integrity and Credibility
The integrity and credibility of this dataset present several challenges that must be acknowledged. One significant concern is the limited sample size of only 30 users, which may lead to sampling bias. Without demographic information, it is difficult to determine whether this sample accurately represents the broader population of Fitbit users. This lack of representation can affect the generalizability of any findings derived from the analysis.

Additionally, the dataset is not current, reflecting data collected during a survey conducted over a two-month period from December 3, 2016, to December 5, 2016. This temporal limitation may result in insights that are less applicable to current trends in health and fitness behaviors.

Given these constraints, our case study will adopt an operational approach, focusing on actionable insights within the defined context rather than making broad generalizations. The company also intends to incorporate external data sources to enrich the analysis; however, due to existing limitations, we may not be able to fulfill all data integration requests. Recognizing these issues is crucial for ensuring transparency and setting realistic expectations for the analysis outcomes.

### 3.4  Relevance to Analysis

### 4.3 Data Cleaning and Transformation
#### Data Cleaning
- Removed duplicates from the dailyActivity_merged to ensure unique records.
- Standardized the date format to ensure consistency across the dataset.
#### Data Transformation
- Concatenated two time periods in the dailyActivity_merged dataset.
- Added a new column weekDay to both dailyActivity_merged and hourlyIntensities_merged dataframes for identifying name of the week.
- Extracted the time from the ActivityHour column and created a new column named time.
- Joined the dailyActivity_merged and sleepDay_merged dataframes to combine activity and sleep data.

#### 4.3.1 Concatenating Data for Two Time Periods

```python
dailyActivity_merged = pd.concat([dailyActivity_merged1, dailyActivity_merged2] , keys = ['Mar-Apr', 'Apr-May']).reset_index(drop = True)
hourlySteps_merged = pd.concat([hourlySteps_merged1, hourlySteps_merged2] , keys = ['Mar-Apr', 'Apr-May']).reset_index(drop = True)
hourlyIntensities_merged = pd.concat([hourlyIntensities_merged1, hourlyIntensities_merged2] , keys = ['Mar-Apr', 'Apr-May']).reset_index(drop = True)
```

#### 4.3.2 Date Conversion

```python
# convert ActivityDate column to Specified date format
dailyActivity_merged['ActivityDate'] = pd.to_datetime(dailyActivity_merged['ActivityDate'], format = '%m/%d/%Y')
```

#### 4.3.4 Null Value Check

```python
# Check for the null values 
dailyActivity_merged.isnull().sum()
```

#### 4.3.4 Duplicate Value Check

```python
# Check for the duplicates
dailyActivity_merged[dailyActivity_merged.duplicated(subset = ['Id', 'ActivityDate'], keep = False)].sort_values(by = 'Id').head()
```

#### 4.3.5 Remove Duplicates

```python
# Group the data to retain the maximum activity level for each common date between the two data frames, 
# treating the maximum activity level as the record for the entire day.

dailyActivity_merged = dailyActivity_merged.groupby(['Id', 'ActivityDate']).agg(
    {
    'TotalSteps': 'max',
    'TotalDistance': 'max',
    'TrackerDistance': 'max',
    'LoggedActivitiesDistance': 'max',
    'VeryActiveDistance': 'max',
    'ModeratelyActiveDistance': 'max',
    'LightActiveDistance': 'max',
    'SedentaryActiveDistance': 'max',
    'VeryActiveMinutes': 'max',
    'FairlyActiveMinutes': 'max',
    'LightlyActiveMinutes': 'max',
    'SedentaryMinutes': 'max',
    'Calories': 'max'
    }
).reset_index()
```

## 5. Analyze Phase

### 5.1 Proportion of time spent in each Intensity level

```python
categorical_intensity = dailyActivity_merged[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']].sum()
categorical_intensity
```

```python
total_intensities_minutes = categorical_intensity[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']].sum()
total_intensities_minutes
```

```python
proportion_minutes_intensities = categorical_intensity[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']].div(total_intensities_minutes)
proportion_minutes_intensities
```

```python
label = ['VeryActive', 'FairlyActive', 'LightlyActive', 'Sedentary']
colors = ['#FF0000', '#FF7F50', '#1E90FF', '#32CD32']
plt.pie(proportion_minutes_intensities, labels = label, autopct = '%1.1f%%', colors = colors)
plt.axis('equal')
plt.title('Proportion of Minutes in Each Intensity level')
plt.show()
```

### 5.2 Proportion of Time Spent in Each Intensity level for different weekdays

```python
dailyActivity_merged['weekDay'] = dailyActivity_merged['ActivityDate'].dt.day_name()
```

```python
minutesSpentActivity_weekDay = dailyActivity_merged.groupby('weekDay').agg({
    'VeryActiveMinutes':'sum',
    'FairlyActiveMinutes':'sum',
    'LightlyActiveMinutes':'sum',
    'SedentaryMinutes':'sum'
}).reset_index()
```

```python
minutesSpentActivity_weekDay
```

```python

activities = ['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']

colors = ['#FF0000', '#FF7F50', '#1E90FF', '#32CD32'] 
N = len(minutesSpentActivity_weekDay['weekDay'])
ind = np.arange(N) 

width = 0.2

fig, ax = plt.subplots(figsize=(10, 6))

for i, activity in enumerate(activities):
    ax.bar(ind + i * width, minutesSpentActivity_weekDay[activity], width, label=activity, color=colors[i])

ax.set_xlabel('Weekday')
ax.set_ylabel('Minutes')
ax.set_title('Intensity level Activity for Each Weekday')
ax.set_xticks(ind + width * 1.5)  
ax.set_xticklabels(minutesSpentActivity_weekDay['weekDay']) 

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

- **Fitbit users spends most time(minutes) in sedentary activities and very less engagement in FairlyActive and VeryActive activities**
- **Across all the weekdays users spent largest portion of time being inactive**

### 5.3 Average Steps on Each Week Day

```python
Steps_Per_Day = dailyActivity_merged.groupby('weekDay').agg({
    'TotalSteps':'mean'
}).reset_index()
```

```python
Steps_Per_Day
```

### 5.4 Average Sleep Hour per Each Week Day

```python
sleepDay_merged['SleepDay'] = pd.to_datetime(sleepDay_merged['SleepDay'], format = '%m/%d/%Y %I:%M:%S %p')
sleepDay_merged['weekDay'] = sleepDay_merged['SleepDay'].dt.day_name()
sleepDay_merged['TotalMinutesAsleep'] = sleepDay_merged['TotalMinutesAsleep']/60
sleepDay_merged['TotalTimeInBed'] = sleepDay_merged['TotalTimeInBed']/60
sleepDay_merged.head()
```

```python
Average_sleep = sleepDay_merged.groupby('weekDay').agg({
    'TotalMinutesAsleep':'mean',
    'TotalTimeInBed':'mean'
}).reset_index()
Average_sleep
```

```python
weekdays_order = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

sns.set(style="whitegrid")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

sns.barplot(x='weekDay', y='TotalSteps', data=Steps_Per_Day, ax=ax1,
            hue='weekDay', palette='Blues', errorbar=None, legend=False, order=weekdays_order)
ax1.axhline(y=7500, color='r', linestyle='--', label='7500 Steps')
ax1.set_xlabel('Weekday')
ax1.set_ylabel('Total Steps')
ax1.set_title('Total Steps Per Day', fontsize=16)
ax1.legend()

sns.barplot(x='weekDay', y='TotalTimeInBed', data=Average_sleep, ax=ax2,
            hue='weekDay', palette='Greens', errorbar=None, legend=False, order=weekdays_order)
ax2.axhline(y=8, color='b', linestyle='--', label='8 Hours')
ax2.set_xlabel('Weekday')
ax2.set_ylabel('Sleep Hours')
ax2.set_title('Sleep Hours per Day', fontsize=16)
ax2.legend()

plt.tight_layout()
plt.show()
```

- **On Friday, Sunday, and Thursday, the total steps are below the target of 7500 steps**.
- **Users get enough sleep on Sunday but fail to reach the recommended 8 hours on other days**.

### 5.5 Correlation Between TotalMinutesAsleep and TotalMinutesSleep

```python
activity_sleep_merged = pd.merge(dailyActivity_merged, sleepDay_merged, how = 'inner', left_on = ['Id', 'ActivityDate'], right_on = ['Id', 'SleepDay'])
activity_sleep_merged.head()
```

```python
sns.set(style='whitegrid')

plt.figure(figsize=(8, 6))
sns.regplot(x='TotalSteps', y='TotalMinutesAsleep', data=activity_sleep_merged, 
            scatter_kws={'s': 100}, 
            line_kws={'color': 'red', 'label': 'Trend Line'},
            scatter=True)

plt.title('Total Steps vs TotalMinutesAsleep', fontsize=16)
plt.xlabel('Total Steps', fontsize=12)
plt.ylabel('Total Minutes Asleep', fontsize=12)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.legend(title='Trend Line', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

- **As people walks more less minutes they asleep means that they sleep more.**

### 5.6 Intensity by WeekDay and Time

```python
hourlyIntensities_merged.head()
```

```python
hourlyIntensities_merged['ActivityHour'] = pd.to_datetime(hourlyIntensities_merged['ActivityHour'], format = '%m/%d/%Y %I:%M:%S %p')
```

```python
hourlyIntensities_merged['weekDay'] = hourlyIntensities_merged['ActivityHour'].dt.day_name()
hourlyIntensities_merged['time'] = hourlyIntensities_merged['ActivityHour'].dt.time
```

```python
DayWise_Intensity = hourlyIntensities_merged.groupby('weekDay')['TotalIntensity'].sum()
```

```python
DayWise_Intensity = DayWise_Intensity.to_frame(name='TotalIntensity')
DayWise_Intensity.reset_index(inplace=True)
DayWise_Intensity.columns = ['weekDay', 'TotalIntensity']
DayWise_Intensity
```

```python
weekday_order = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))  # Set figure size
bar_plot = sns.barplot(x='weekDay', y='TotalIntensity', data=DayWise_Intensity,
                       hue='weekDay', palette='viridis', edgecolor='black',
                       order=weekday_order)  # Use order parameter to sort by weekdays

bar_plot.set_title('Total Intensity by Weekday', fontsize=16, fontweight='bold')
bar_plot.set_xlabel('Weekday', fontsize=12)
bar_plot.set_ylabel('Total Intensity', fontsize=12)

for p in bar_plot.patches:
    bar_plot.annotate(format(int(p.get_height()), ','),  # Convert height to int
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha='center', va='center', 
                      xytext=(0, 9), textcoords='offset points')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- **Saturday has the highest total intensity, suggesting users are most active on weekends.**
- **Friday shows the minimum total intensity, suggesting users are less active at the end of workweek.**

```python
hourly_intensity = hourlyIntensities_merged.groupby('time')['AverageIntensity'].mean()
hourly_intensity = hourly_intensity.to_frame(name='AverageIntensity')
hourly_intensity.reset_index(inplace=True)
hourly_intensity.columns = ['time', 'AverageIntensity']
hourly_intensity
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))  # Set figure size
bar_plot = sns.barplot(x='time', y='AverageIntensity', data=hourly_intensity,hue = 'time', palette='viridis', edgecolor='black', legend = False)
bar_plot.set_title('Average Intensity by Time', fontsize=16, fontweight='bold')
bar_plot.set_xlabel('Time of Day', fontsize=12)
bar_plot.set_ylabel('Average Intensity', fontsize=12)

for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.3f'),  # Show height with three decimal places
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha='center', va='center', 
                      xytext=(0, 5), textcoords='offset points')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- **Users are most active on day time and the intensity reach to peak between 5PM to 7PM**

### 5.7 Correlation between total distance and calories

```python
sns.set(style='whitegrid')

plt.figure(figsize=(8, 6))

sns.scatterplot(x='TotalDistance', y='Calories', data=dailyActivity_merged, 
                s=100, color='blue', label='Data Points')

z = np.polyfit(dailyActivity_merged['TotalDistance'], dailyActivity_merged['Calories'], 1)
p = np.poly1d(z)

plt.plot(dailyActivity_merged['TotalDistance'], p(dailyActivity_merged['TotalDistance']), 
         color='red', linestyle='--', label='Trend Line')

plt.title('Total Distance vs Calories', fontsize=16)
plt.xlabel('Total Distance Walked', fontsize=12)
plt.ylabel('Calories Burned', fontsize=12)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.legend(title='Legend', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

plt.show()
```

- **The more distance users cover, the more calories they burn.**
